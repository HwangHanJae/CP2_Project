# 데이터 정제

해당 데이터의 시간은 UTC+4 이다.

따라서 데이터의 event_time을 4를 더한 시간으로 변경해준다.

```python
# gmt 날짜 저장을 위한 별도 리스트 생성
gmt_list = []
for time in copy_df['event_time']:
    gmt_list.append(time[:19])
gmt_list
```

```python
copy_df['gmt_time'] = gmt_list
```

```python
copy_df['gmt_time'] = pd.to_datetime(copy_df['gmt_time'])
type(copy_df['gmt_time'][0])
```

```python
copy_df['gmt_time'] = copy_df['gmt_time'] + timedelta(hours=4)
copy_df.head()
```

```python
copy_df.drop(['event_time'], axis=1, inplace=True)
copy_df
```

# 유저, 제품별 세션 체류시간 계산

문제점
- 사용자가 이탈하는 경우 세션 로그가 남지 않기 때문에 마지막 페이지의 체류 시간을 알 수 없고, 사이트 체류 시간 계산에 포함되지 않는다.
- ex) view를 1회만 한 경우 -> 체류 시간 없음. view를 3회 한 경우 -> 두 번째 체류 시간까지만 계산됨


하지만 사이트 체류 시간이 유의미하지 않은 지표는 아님
- 수많은 사용자들이 수많은 페이지를 빈번하게 오고가는 상황 속에 모든 로그에서 동일하게 마지막 페이지의 체류 시간을 계산하지 않음
- 즉, 모두가 일관된 기준으로 측정함
- 또한 데이터의 양이 상당하기 때문에 한 개 페이지의 시간 차이는 크게 사용자들의 트렌드를 반영하지 못한다고 봐도 옳을 것

```python
# 세션, 제품을 기준으로 groupby 진행
'''
동일한 세션에 대해

- 가장 마지막(최대)의 시간을 값으로 하는 그룹
- 가장 처음(최소)의 시간을 값으로 하는 그룹

각각 생성 후 데이터프레임으로 변환
'''
max_grouped = copy_df.groupby(['user_session', 'user_id', 'product_id'])['gmt_time'].max()
min_grouped = copy_df.groupby(['user_session', 'user_id', 'product_id'])['gmt_time'].min()
```

```python
# 가장 최근 접속 시간
max_time_df = pd.DataFrame(max_grouped)
max_time_df.columns = ['latest']
max_time_df
```

```python
# 처음 접속 시간
min_time_df = pd.DataFrame(min_grouped)
min_time_df.columns = ['start']
min_time_df
```

```python
time_df = pd.concat([max_time_df, min_time_df], axis=1)
time_df
```

```python
time_df['session_time'] = time_df['latest'] - time_df['start']
time_df
```

```python
# 세션 타임이 긴 사용자
time_df.sort_values(by=['session_time'], ascending=False)
```

## 이상치 제거
- 세션 타임이 1일 이상: 이상치로 판단 -> 제거
- 세션 타임이 0초 -> 이상치는 아니지만 의미없는 데이터로 판단 -> 제거

```python
# 세션 타임이 1일 이상인 데이터 제거
drop_session_time = time_df.loc[time_df['session_time'].dt.days >= 1, 'session_time']
```

```python
time_df = time_df.loc[-time_df['session_time'].isin(drop_session_time)].reset_index()

time_df.sort_values(by=['session_time'], ascending=False)
```

```python
# 세션 타임이 0초인 데이터 제거
drop_session_time_2 = time_df.loc[time_df['session_time'].dt.seconds == 0, 'session_time']
```

```python
time_df = time_df.loc[-time_df['session_time'].isin(drop_session_time_2)].reset_index()

time_df.sort_values(by=['session_time'], ascending=False)
```

# 유저의 제품별 체류 시간은?

```python
# 유저의 제품별 체류 시간 합계는?
user_product_time_sum = time_df.groupby(['user_id','product_id'])['session_time'].sum()
user_product_time_sum = pd.DataFrame(user_product_time_sum).reset_index()
user_product_time_sum
```

```python
# 유저의 제품별 체류 시간 평균은?
user_product_time_mean = time_df.groupby(['user_id','product_id'])['session_time'].mean()
user_product_time_mean = pd.DataFrame(user_product_time_mean).reset_index()
user_product_time_mean
```

```python
# 유저의 제품별 평균 체류 시간은?
user_time = pd.DataFrame(user_product_time_sum.groupby('user_id')['session_time'].mean()).reset_index()
user_time
```

```python
user_time['session_time'].mean()
```

view만 한 유저의 평균 체류 시간은 9분임을 알 수 있음

## 컬럼 분석 기준

전환이 일어나지 않은 유저에게 체류 시간은..?
- 체류 시간이 짧다: 구매 확실히 안 할 것 => 추천해봤자 의미 없음
- 체류 시간이 보통: 구매를 할지도 안 할지도 모름(잠재고객) => 그렇담 이 제품은 어때? 하고 비슷한 상품 추천
- 체류 시간이 길다: 구매를 고려하고 있음 => 다른 사람들은 이 제품을 얼마나 샀는지 보여주면 되지 않을까

```python
# 파생변수 생성

'''
score : 유저의 체류 시간을 기준으로 유저의 제품에 대한 선호도 점수를 매긴 변수
- 1 : 체류 시간 1초 이하 && 1시간(3600초) 이상
- 2 : 체류 시간 2초 이상 1분(60초) 미만 && 15분(900초) 이상 1시간(3600초) 미만
- 3 : 체류 시간 1분 이상 15분(900초) 미만
'''

#user_time['score'] = np.where(user_time['session_time'].dt.seconds <= 1, 1)

score=[]                              
for i in user_product_time_sum['session_time'].dt.seconds:
    if i <= 1:
        score.append(1)
    elif (i >= 2) & (i < 60):
        score.append(2)
    elif (i >= 60) & (i < 900):
        score.append(3)
    elif (i >= 900) & (i < 3600):
        score.append(2)
    else:
        score.append(1)
```

```python
user_product_time_sum['score'] = score
user_product_time_sum.head()
```
