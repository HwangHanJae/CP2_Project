```python
import pandas as pd
```

```python
df=pd.read_excel('https://github.com/springcoil/marsmodelling/blob/master/relay-foods.xlsx?raw=true', 
                 sheet_name='Purchase Data - Full Study')
df.head()
```

주어진 데이터를 확인해보면 하나의 행에 UserId(유저아이디)와 OrderDate(주문날짜)가 함께 구성되어 있습니다.  
- 이것은 코호트의 기본 전제인 (동일한 경험, 동일한 기간)을 충족하는 데이터로 전처리를 할 수 있음

**우리 데이터에 접목하면, 유저ID,동일하게 방문한 날짜를 사용할 수도 있을 것 같음!**

```python
df.info()
```

결측치는 없습니다.

# 분석에 불필요한 컬럼 제거

```python
drop_columns = ['CommonId','PupId','PickupDate','TotalCharges']
df = df.drop(drop_columns, axis=1)
df.head()
```

# 주문일자(OrderDate) 칼럼으로 단위 기간 만들기
- 코호트분석시 데이터를 집계하는 단위는 **월 단위**로 진행합니다.
- **하지만 우리 데이터는 10월 한달의 데이터이기 때문에 일단위 혹은 주 단위로 진행해야 할 것입니다.**

```python
df['OrderDate'].head() 
```

```python
df['OrderDate'].dtype
```

주문일자(OrderDate)의 dtype을 확인해보면 시계열 데이터임을 확인할 수 있습니다.

```python
df['OrderPeriod'] = df['OrderDate'].dt.strftime('%Y-%m')
df.head()
```

# 유저의 첫번째 주문을 토대로 유저별 코호트 그룹 설정하기
- 코호트 분석의 전제조건 중 동일한 경험을 만드는 과정입니다.
- 실습데이터에서의 동일한 경험은 첫번째 주문입니다.
- **우리 데이터의 동일한 경험은 10월 첫번째 방문이라고 가정할 수 있음**


```python
df = df.set_index('UserId')
df.head()
```

```python
first_order = df.groupby(df.index)['OrderDate'].min()
first_order.head()
```

각 유저(UserId)별 첫번째 주문 일자를 추출합니다. 

```python
first_order = first_order.dt.strftime('%Y-%m')
first_order.head()
```

first_order(첫번째 주문)에서 년-월만 추출합니다.

```python
df['FirstOrder'] = first_order
df = df.reset_index()
df.head()
```

FirstOrder(첫번째 주문일자)를 추가하고 reset_index()를 통하여 다시 데이터프레임을 본래의 형태로 바꾸어 줍니다.

# 코호트 그룹과 주문기간으로 데이터 집계하기

```python
grouped = df.groupby(['FirstOrder','OrderPeriod'])
```

FirstOrder(첫 주문일자)와 OrderPeriod(주문기간)을 기준으로 그룹화 합니다.

```python
cohorts = grouped['UserId'].nunique()
cohorts.head()
```

grouped 객체에 UserId(유저 아이디)를 집계하며,   
집계함수로 각 unique값의 개수를 세는 nunqiue함수를 사용합니다.  
이렇게 하면 고객이 여러번 방문했다 하더라도 한 번으로 집계하게 되어 월별 순방문자수를 파악할 수 있습니다.

```python
cohorts = cohorts.reset_index()
cohorts= cohorts.rename({'UserId':'TotalUsers'}, axis=1)
cohorts.head()
```

UserId 컬럼을 의미에 맞게 TotalUsers로 변경합니다.

# 데이터가 없는 기간 체크 및 데이터 전처리

image.png

코호트 분석 화면을 보면 기간을 표현하는 칼럼(x축)은 0부터 9까지 표기 되어 있습니다. 

고객이 언제 구매하였던 간에 처음 들어온 기간은 0, 두 번째 기간은 1, 세 번째 기간은 2, 네 번째 기간은 3으로 표기합니다.


```python
each_period = cohorts['FirstOrder'].value_counts().sort_index()
each_period
```

그런데 첫 번째 코호트 기간은 15번으로 구성되어 있기 때문에 두 번째 코호트 기간은 15-1= 14여야 하는데 13인것을 확인할 수 있습니다.

그 외의 데이터를 살펴보면 13,12,11,10... 등 정상적으로 처리되어 있는 것을 확인할 수 있습니다.

```python
cohorts[cohorts['FirstOrder']=='2009-02']
```

OrderPeriod가 2010-03인 데이터가 없는 것을 확인할 수 있습니다.

```python
cohorts.loc[cohorts.shape[0], 'FirstOrder'] = '2009-02'
cohorts.loc[cohorts.shape[0]-1,'OrderPeriod']='2010-03'
cohorts.loc[cohorts.shape[0]-1,'TotalUsers']=0
cohorts = cohorts.sort_values(['FirstOrder','OrderPeriod'])
```

```python
cohorts[cohorts['FirstOrder']=='2009-02']
```

```python
cohorts['FirstOrder'].value_counts()
```

데이터가 올바르게 들어간것을 확인할 수 있습니다.

```python
each_period = cohorts['FirstOrder'].value_counts().sort_index()
each_period
```

# 기간별 레이블 붙이기
- 코호트 분석에서 처음에 들어온 기간은 0, 두 번째 기간은 1, ...  표기가 됩니다.
- 실습데이터에도 같은 레이블을 붙입니다.

```python
cohort_period = []
for x in each_period:
  for y in range(x):
    cohort_period.append(y)
```

```python
cohorts['CohortPeriod'] = cohort_period
cohorts.head(20)
```

```python
cohorts = cohorts.set_index(['FirstOrder','CohortPeriod'])
cohorts.head()
```

FirstOrder(첫번째 주문일자)와 CohortPeriod를 기준으로 인덱스를 만들어 줍니다.

```python
cohorts = cohorts['TotalUsers'].unstack(1)
cohorts
```

unstack(1)를 사용하여 멀티인덱스 중 두 번째 인덱스(CohortPeriod)가 칼럼으로 변경되고,  
결과 y축은 코호트, x축은 기간으로 표현되는 코호트 분석의 형태를 갖게 만듭니다.

# 코호트별 사용자 유지율 구하기
- 코호트분석은 숫자가 아닌 비율로 표현되는 경우가 많습니다.
- 여기서 **비율**은 처음의 순 방문자에 비해 각 기간의 순 방문자가 얼마나 변화하였는지에 대한 것 입니다.


```python
cohorts[0]
```

cohorts[0]은 각 코호트의 첫 번째 기간(0)에 대한 데이터로 구성되어 있습니다.

```python
user_retention = cohorts.divide(cohorts[0], axis=0)
user_retention
```

# 코호트 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
plt.figure(figsize=(12,8))
plt.title('Cohort Analysis')
sns.heatmap(user_retention, annot=True, fmt='.0%')
plt.show()
```
